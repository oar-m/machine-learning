# 2変数正規分布の乱数生成

```python
from numpy.random import multivariate_normal
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np

 
def cal_random(mean, sigma, data_num):
    values = multivariate_normal(mean, sigma, data_num)
    # plot
    sns.jointplot(values[:,0], values[:,1])
    plt.show()
```

```python
mu = [0, 0]
sigma = [[30, 20], [20, 50]]
data_num = 1000
cal_random(mu, sigma, data_num)

mu = [0, 10000]
sigma = [[30, 20], [20, 50]]
cal_random(mu, sigma, data_num)

mu = [0, 0]
sigma = [[15, 10], [10, 15]]
cal_random(mu, sigma, data_num)
```

# 標準化
- 説明変数のスケールが揃う
- 平均値０
- 分散1
- 分散共分散行列の共分散が相関係数に
- 当たり前だがこれらはデータ数が多くなると理論値と実測値が近く

```python
def standard_with_data(data):
    mean = np.mean(data, axis=0)
    std = np.std(data, axis=0)
    norm_data = (data - mean) / std
    print("From data")
    print(f"variance and covariance of matrix \n> {np.cov(norm_data, rowvar=0, bias=1)}")
    print(f"mean > {np.mean(norm_data, axis=0)}\n")
    return norm_data

def standard_with_param(data, mean, sigma):
    length = len(sigma[0])
    std = [np.sqrt(sigma[i][i]) for i in range(length)]
    # print(std)
    norm_data = (data - mean) / std
    print("From param")
    print(f"variance and covariance of matrix \n> {np.cov(norm_data, rowvar=0, bias=1)}")
    print(f"mean > {np.mean(norm_data, axis=0)}\n")
    return norm_data
                    
def cal_normaliz_random(mean, sigma, data_num):
    values = multivariate_normal(mean, sigma, data_num)
    std_data_data = standard_with_data(values)
    std_data_param = standard_with_param(values, mean, sigma)

    sns.jointplot(std_data_data[:,0], std_data_data[:,1])
    sns.jointplot(std_data_param[:,0], std_data_param[:,1])
    plt.show()
```

```python
print("Normal")
mu = [0, 0]
sigma = [[30, 20], [20, 50]]
data_num = 1000
cal_normaliz_random(mu, sigma, data_num)

print("Focus on the mu")
mu = [0, 10000]
sigma = [[30, -20], [-20, 50]]
cal_normaliz_random(mu, sigma, data_num)

print("Focus on the sigma  -> use same variance")
mu = [0, 0]
sigma = [[15, 10], [10, 15]]
cal_normaliz_random(mu, sigma, data_num)

print("Focus on the sigma  -> covariance = 0")
mu = [0, 0]
sigma = [[10, 0], [0, 10]]
cal_normaliz_random(mu, sigma, data_num)

print("Focus on the data number  -> data number 10times")
mu = [0, 0]
sigma = [[30, 20], [20, 50]]
data_num = 10000
cal_normaliz_random(mu, sigma, data_num)
```

# 無相関化
- 分散共分散行列が対角化される　→ 　相関係数が0

```python
def cal_decorrelation_random(mean, sigma, data_num):
    values = multivariate_normal(mean, sigma, data_num)
    
    # 固有値、固有ベクトル (回転行列)
    _, S = np.linalg.eig(sigma)

    # 回転行列を使ってデータを線形変換
    values = np.dot(S.T, values.T).T
    print(values.T.shape)
    print(f"Correlation coefficient \n> {np.corrcoef(values.T)}")

    # plot
    sns.jointplot(values[:,0], values[:,1])
    plt.show()
```

```python
print("Normal")
mu = [0, 0]
sigma = [[30, 20], [20, 50]]
data_num = 1000
cal_decorrelation_random(mu, sigma, data_num)

print("Focus on the mu")
mu = [0, 100]
sigma = [[30, 20], [20, 50]]
cal_decorrelation_random(mu, sigma, data_num)

print("Focus on the sigma  -> use same variance")
mu = [0, 0]
sigma = [[11, 10], [10, 11]]
cal_decorrelation_random(mu, sigma, data_num)

print("Focus on the data number  -> data number 10 times")
mu = [0, 0]
sigma = [[30, 20], [20, 50]]
data_num = 10000
cal_decorrelation_random(mu, sigma, data_num)
```
